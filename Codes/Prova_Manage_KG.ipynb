## Prova con OWL file

```python
from rdflib import Graph
import networkx as nx

# Leggi il file OWL in un grafo RDF utilizzando rdflib
file_path = "bionutrition_cam9.owl"
g = Graph()
g.parse(file_path, format="xml")  # Assicurati di specificare il formato corretto del tuo file OWL

# Costruisci un grafo NetworkX basato sui dati RDF
import networkx as nx

# Crea un nuovo grafo NetworkX
G1 = nx.MultiDiGraph()  # Usa MultiDiGraph se il tuo grafo RDF contiene archi multipli tra gli stessi nodi


```

```python
# Estrai i nodi e gli archi dal grafo RDF e aggiungili al grafo NetworkX
for subject, predicate, obj in g:
    G1.add_node(subject)
    G1.add_node(obj)
    G1.add_edge(subject, obj, key=predicate)

G1
```

```python
print(G1.nodes)
```

```python
# Itera sui triple nel grafo RDF
for predicate in G1:
    print(f"Predicate: {predicate}")
```

## Knowledge linker on local KG
* https://stackoverflow.com/questions/52046394/how-to-perform-entity-linking-to-local-knowledge-graph

```python
input = 'new york is the big apple'.split()


def spans(lst):
    if len(lst) == 0:
        yield None
    for index in range(1, len(lst)):
        for span in spans(lst[index:]):
            if span is not None:
                yield [lst[0:index]] + span
    yield [lst]

knowledgebase = [
    ['new', 'york'],
    ['big', 'apple'],
]

out = []
scores = []

for span in spans(input):
    score = 0
    for candidate in span:
        for uid, entity in enumerate(knowledgebase):
            if candidate == entity:
                score += 1
    out.append(span)
    scores.append(score)

leaderboard = sorted(zip(out, scores), key=lambda x: x[1])

for winner in leaderboard:
    print(winner[1], ' ~ ', winner[0])
```
